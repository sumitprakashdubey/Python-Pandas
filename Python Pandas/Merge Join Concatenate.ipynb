```python
left = pd.DataFrame({
   'id':[1,2,3,4,5,6],
   'Name': ['Sumit', 'Amit', 'Andrew', 'Alex', 'Atul','Shyam'],
   'Job_id':['job1','job2','job3','job4','job5','Job6']})
right = pd.DataFrame(
   {'id':[1,2,3,4,5,7],
   'Name': ['Bikash', 'Arup', 'Brayan', 'Lee', 'Xavier','Richa'],
   'Job_id':['job2','job4','job3','job6','job5','job5']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on="id")
```

```python
pd.merge(left, right, on=["id","Job_id"])
```

```python
pd.merge(left, right, on=["id"],how='inner')
```

```python
pd.merge(left, right, on=["id"],how='outer')
```

```python
pd.merge(left, right, on=["id"],how='left')
```

```python
pd.merge(left, right, on=["id"],how='right')
```

```python
pd.merge(left, right,left_on="id",right_on="id",how='inner')
```

```python
pd.merge(left, right,left_index=True,right_index=True,how='outer')
```

```python
pd.merge(left, right, on="id", how="inner", validate="one_to_one")
```

```python
pd.merge(left, right, on="id", how="inner", validate="one_to_many")
```

```python
pd.merge(left, right, on="id", how="inner", validate="many_to_many")
```

```python
left.join(right,lsuffix='A',how='outer')
```
